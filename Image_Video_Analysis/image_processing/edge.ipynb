```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
image_rgb = matplotlib.image.imread('img.jpg')
image_rgb.shape
```

```python
plt.imshow(image_rgb)
```

```python
def apply_filter(image, filter):
    new_image = np.zeros(image.shape)
    for i in range(1, image.shape[0] - 1):
        for j in range(1, image.shape[1] - 1):
            new_image[i, j] = np.sum(image[i - 1:i + 2, j - 1:j + 2] * filter)
    return new_image

```

```python
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
plt.imshow(apply_filter(image_gray, np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])), cmap='gray')
```

```python
plt.imshow(apply_filter(image_gray, np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])), cmap='gray')
```

```python
image_person = matplotlib.image.imread('person.jpg')
plt.imshow(image_person)
```

```python
image_person_gray = cv2.cvtColor(image_person, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(apply_filter(image_person_gray, np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])), cmap='gray')
```

```python
plt.imshow(apply_filter(image_person_gray, np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])), cmap='gray')
```
