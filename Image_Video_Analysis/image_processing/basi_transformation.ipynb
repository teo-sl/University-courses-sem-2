```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = np.array(Image.open('img.jpg').convert('L'))
img.shape
```

Define basic transformation using:

$$x' = Ax$$

where A is a 2x3 matrix and x is a 3x1 vector ([x, y, 1], we're using omogenous coordinates).

Changing A we can obtain different transformations.


```python
def apply_transformation(img,matrix):
    new_image = np.zeros((img.shape[0]+2000,img.shape[1]+2000))
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            k,l = np.dot(matrix,np.array([[i],[j],[1]]))
            k,l = int(k),int(l)
            new_image[k,l] = img[i,j]
    return new_image
```

```python
theta = np.pi/4
m1 = [
    [np.cos(theta), -np.sin(theta), 1500],
    [np.sin(theta),np.cos(theta), 1000]
]
m1 = np.array(m1)
new_img = apply_transformation(img,m1)
```

```python
plt.imshow(new_img[300:2400,900:3000],cmap='gray')
```

```python

m1 = [
    [0.8, 0.3, 3],
    [0.3, 0.2, 6]
]
m1 = np.array(m1)
new_img = apply_transformation(img,m1)
plt.imshow(new_img,cmap='gray')
```
