```python
import numpy as np
import networkx as nx
```

Dimostriamo che la katz centrality pu√≤ essere ottenuta dalla regular equivalence #1 andando a sommare per tutti quei nodi adiacenti al nodo target.

```python
def katz_centrality(A,alpha):
    M = A.shape[0]
    katz = (np.linalg.inv(np.eye(M)-alpha*A.T) - np.eye(M)) @ np.ones(M)
    return katz


def regular_equivalence(A,alpha,e):
    M = A.shape[0]
    return np.linalg.inv((np.eye(M)-alpha*A.T)) @ e
    
```

```python
np.random.seed(0)
A = np.random.randint(0,2,(5,5))


print(katz_centrality(A,0.85))

print(regular_equivalence(A,0.85,np.ones(5)))
```

```python
G = nx.from_numpy_array(A, create_using=nx.DiGraph)

# compute katze centrality
print(nx.katz_centrality_numpy(G, alpha=0.85,beta=1,normalized=False))
```
