```python
import networkx as nx
import numpy as np
import plotly.express as px
```

```python
#Â create graph according to the preferential attachment model
G = nx.barabasi_albert_graph(10000, 100)
```

```python
# plot the degree distribution
degrees = [G.degree(n) for n in G.nodes()]
px.scatter(x=range(len(degrees)), y=degrees, log_y=True,log_x=True)
```

```python
# get the local clustering coefficient for each node
clustering = nx.clustering(G)
# get the degree for each node
degrees = [G.degree(n) for n in G.nodes()]
```

```python
# plot the degree vs clustering coefficient
px.scatter(x=degrees, y=clustering, log_y=True, log_x=True)
```
